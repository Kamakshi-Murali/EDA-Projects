```python
df['Humidity(%)'] = df['Humidity(%)'].astype('float')
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
plt.plot(df['Hour'], df['Rented Bike Count'])
plt.title('Bikes rented per Hour')
plt.xlabel('Hours')
plt.ylabel('Rented Bike Count')
plt.grid(True)
plt.xticks(df['Hour'])
plt.yticks([500,1000,1500,2000,2500,3000,3500])
plt.show()
```

```python
# Bar Chart
plt.bar(df['Hour'], df['Rented Bike Count'])
plt.title('Bar Chart')
plt.xlabel('Hour')
plt.ylabel('Rented Bike Count')
plt.grid(True)
plt.xticks(df['Hour'])
plt.yticks([500,1000,1500,2000,2500,3000,3500])
plt.show()
```

```python
plt.hist(df['Rented Bike Count'], bins = 5)
plt.title('Histograph', color='#FF5733')
plt.xlabel('Rented Bike Count')
plt.ylabel('Y-Axis')
plt.grid(True)
plt.show()
```

```python
# Pie Chart
plt.pie(df['Rented Bike Count'], labels = df['Seasons'] , autopct = '%1.1f%%')
plt.title('Pie Chart')

plt.show()
```

```python
# Scatter Plot
plt.scatter(df['Hour'], df['Rented Bike Count'])
plt.title('Scatter Plot')
plt.show()
```

```python
# Seaborn charts

sns.lineplot(x=df['Hour'], y=df['Rented Bike Count'], data=df)
plt.title('Line plot')
plt.show()
```

```python
sns.scatterplot(x=df['Hour'], y=df['Rented Bike Count'], data=df)
plt.title('Hour vs Rented Bike Count')
plt.show()
```

```python
sns.barplot(x=df['Hour'], y=df['Rented Bike Count'], data=df)
plt.title('Bar chart')
plt.show()
```

```python
sns.histplot(x = df['Rented Bike Count'], data = df, bins = 5, kde = True)  # Bins represent the range of values
plt.title('Histogram of Rented Bike Count')
plt.show()
```

```python
plt.pie(df['Rented Bike Count'],labels=df['Hour'], autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.axis('equal')               
plt.show()
```
